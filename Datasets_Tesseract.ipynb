```python
# Funcion para extraer datos de las cajas 
def ine(image_ine,cor_ine):

  #Carga la image
  img = Image.open(image_ine)

  #Cargar texto 
  path1 = path+'/..'
  with open(cor_ine) as f:
    lines = f.readlines()
  lines[0].split(" ")
  lista=[]
  for line in lines:
    lista.append(line.split(" "))
  
  cor = lista 

  dic={}
  labels=["nombre", "direccion", "curp", "sexo", "fecha de nacimiento"]

  for i,val in enumerate(labels):
    x,y = img.size
    xc = float(cor[i][1])
    yc= float(cor[i][2])
    ancho=float(cor[i][3])
    alto=float(cor[i][4])

    x1=xc-(ancho)/2
    x2=xc+(ancho)/2
    y1=yc-(alto)/2
    y2=yc+(alto)/2 

    a = imagus.crop(img,x1=x1,y1=y1,x2=x2,y2=y2)
    ei = pytesseract.image_to_string(a, lang='spa')
    dic[labels[i]]=ei

  # for j,el in enumerate(dic):
  #   # print(dic[labels[j]])
  return dic
```
